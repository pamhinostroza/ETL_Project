```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
living_cost = "Resources/cost_living_index.csv"
living_cost_df = pd.read_csv(living_cost)
living_cost_df.head()
```

```python
homes = "Resources/median_home_prices.csv"
homes_df = pd.read_csv(homes)
homes_df.head()
```

```python
income = "Resources/median_household_income.csv"
income_df = pd.read_csv(income)
income_df.head()
```

```python
living_cost_df = living_cost_df.drop(columns = ['groceryCost','housingCost','utilitiesCost','transportationCost','miscCost'])
living_cost_df = living_cost_df.rename(columns={'State':'state', 'costIndex':'cost_index'})
living_cost_df.set_index("state", inplace=True)
living_cost_df.head()
```

```python
homes_df = homes_df.rename(columns={'State':'state', 'MedianValue':'median_value'})
homes_df.set_index("state", inplace=True)
homes_df.head()
```

```python
income_df = income_df.drop(columns=['ID Race','ID State','ID Year','Household Income by Race Moe','Slug State'])
income_df = income_df.loc[(income_df['Year'] >= 2018) & (income_df['Race'] == 'Total')]
income_df = income_df.drop(columns=['Race','Year'])
income_df = income_df.rename(columns={'State':'state', 'Household Income by Race':'income'})
income_df.set_index("state", inplace=True)
income_df.head()
```
